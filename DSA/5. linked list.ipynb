```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None
        
class Solution:
    def constructDLL(self, arr):
        if not arr:  # If the array is empty, return None
            return None
        head = Node(arr[0])
        temp_pointer = head
        for i in range(1, len(arr)):
            new_node = Node(arr[i])
            temp_pointer.next = new_node
            new_node.prev = temp_pointer
            temp_pointer = temp_pointer.next
        return head
        
    def display(self, LL):
        temp_pointer = LL
        while temp_pointer:
            print(temp_pointer.data, end=" <-> ")
            temp_pointer = temp_pointer.next
        print("None")
```

```python

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

class Solution:
    # Function to insert a new node after given position p in a doubly linked list.
    def addNode(self, head, p, x):
        new_node = Node(x)

        # Case 1: If the list is empty
        if head is None:
            return new_node  

        temp = head

        # Traverse to position 'p'
        for _ in range(p):
            if temp.next is None:  # If p is out of bounds, insert at the end
                return
            temp = temp.next

        # Case 2: Inserting at the end (if temp.next is None)
        if temp.next is None:
            temp.next = new_node
            new_node.prev = temp
            return head

        # Case 3: Insert new_node after temp (middle insertion)
        new_node.next = temp.next
        new_node.prev = temp
        temp.next.prev = new_node
        temp.next = new_node

        return head  # Return updated head

    # Function to print the linked list
    def printList(self, node):
        if node is None:
            print("List is empty.")
            return
        while node:
            print(node.data, end=" <-> " if node.next else "\n")
            node = node.next

# Example Usage
dll = Solution()
head = Node(6)
head.next = Node(5)
head.next.prev = head
head.next.next = Node(5)
head.next.next.prev = head.next
head.next.next.next = Node(6)
head.next.next.next.prev = head.next.next
head.next.next.next.next = Node(3)
head.next.next.next.next.prev = head.next.next.next
head.next.next.next.next.next = Node(5)
head.next.next.next.next.next.prev = head.next.next.next.next
head.next.next.next.next.next.next = Node(8)
head.next.next.next.next.next.next.prev = head.next.next.next.next.next

print("Original List:")
dll.printList(head)

head = dll.addNode(head, 7, 7)  # Insert 7 at position 7 (out of range, should append at end)
print("\nAfter Inserting 7 at Out-of-Range Position:")
dll.printList(head)
```

```python
class Solution:
    def delete_node(self, head, x):
        # Edge case: If the list is empty
        if not head:
            return None  

        temp_pointer = head

        # Traverse to the x-th node (0-based index)
        for _ in range(1,x):
            if not temp_pointer.next:  # If x is out of bounds
                return head
            temp_pointer = temp_pointer.next

        # Delete the node
        if temp_pointer.prev:
            temp_pointer.prev.next = temp_pointer.next
        if temp_pointer.next:
            temp_pointer.next.prev = temp_pointer.prev

        # If head is being deleted, update the head
        if temp_pointer == head:
            head = temp_pointer.next

        return head  # Return the updated head
```

```python
import time

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

class Solution:
    def Brute_first_reverseDLL(self, head):
        stack = []
        temp = head

        # Push all values into stack
        while temp:
            stack.append(temp.data)
            temp = temp.next

        # Assign values in reverse order
        temp = head
        while temp:
            temp.data = stack.pop()
            temp = temp.next

        return head

    def Optimal_reverse(self, head):
        temp_pointer = head
        new_head = None

        while temp_pointer:
            temp_value = temp_pointer.prev
            temp_pointer.prev = temp_pointer.next
            temp_pointer.next = temp_value

            new_head = temp_pointer
            temp_pointer = temp_pointer.prev

        return new_head

    def printList(self, node, limit=10):
        """Prints up to 'limit' elements for preview."""
        count = 0
        while node and count < limit:
            print(node.data, end=" <-> " if node.next and count < limit - 1 else "\n")
            node = node.next
            count += 1
        if node:
            print("... (truncated)")

# Create a Doubly Linked List from a list of values
def createDLL(values):
    if not values:
        return None

    head = Node(values[0])
    temp = head
    for val in values[1:]:
        new_node = Node(val)
        temp.next = new_node
        new_node.prev = temp
        temp = new_node
    return head

# Large input test (100,000 elements)
values = list(range(1, 100001))  # [1, 2, 3, ..., 100000]
dll = Solution()
head = createDLL(values)

print("Original List (First 10 elements):")
dll.printList(head)

# Measure time for brute force approach
start_time = time.time()
head_brute = dll.Brute_first_reverseDLL(head)
brute_time = time.time() - start_time

print("\nReversed List (Brute Force, First 10 elements):")
dll.printList(head_brute)
print(f"Time taken (Brute Force): {brute_time:.6f} seconds")

# Reset DLL for optimal method
head = createDLL(values)

# Measure time for optimal approach
start_time = time.time()
head_optimal = dll.Optimal_reverse(head)
optimal_time = time.time() - start_time

print("\nReversed List (Optimal Swap, First 10 elements):")
dll.printList(head_optimal)
print(f"Time taken (Optimal Approach): {optimal_time:.6f} seconds")

```

```python
class Solution(object):
    def Using_set_hasCycle(self, head):
        set=set()
        pointer=head
        while pointer:
            if pointer in set:
                return True
            set.add(pointer)
            pointer=pointer.next
        return False
    
    def Using_dict_hasCycle(self, head):
        dict={}
        pointer=head
        while pointer:
            if pointer in dict:
                return True
            dict[pointer]=True
            pointer=pointer.next
        return False
    
    def Using_TurtoiseHare_hasCycle(self, head):
        slow=head
        fast=head
        while fast and fast.next:
            slow=slow.next
            fast=fast.next.next
            if slow==fast:
                return slow
        return None
```

```python
class Solution(object):
    def brute_force_isPalindrome(self, head):
        stack=[]
        temp=head
        count=0
        while temp:
            stack.append(temp.val)
            temp=temp.next
            count=count+1

        temp=head
        mid=count//2
        for _ in range(mid):
            if temp.val!=stack.pop():
                return False
            temp=temp.next
        return True
    
    # Idea: Only store the first half of the list in a stack, then compare while traversing the second half.
    def optimal_isPalindrome(self, head):

        stack = []
        slow = fast = head
        
        while fast and fast.next:
            stack.append(slow.val)
            slow = slow.next
            fast = fast.next.next
        
        if fast: # Odd length, skip the middle element
            slow = slow.next
        
        while slow:
            if slow.val != stack.pop():
                return False
            slow = slow.next
        
        return True
    
```

```python
# Idea: Reverse the second half of the linked list in place and compare with the first half.
def optimal2_isPalindrome(self, head):
        if not head or not head.next:
            return True
        
        slow = fast = head
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next

        prev = None
        while slow:
            temp = slow.next
            slow.next = prev
            prev = slow
            slow = temp
        
        left, right = head, prev
        while right:
            if left.val != right.val:
                return False
            left = left.next
            right = right.next
        
        return True
```

```python
class Solution(object):
    def BruteForce1_getIntersectionNode(self, headA, headB):
        if not headA or not headB:
            return None

        a, b = headA, headB

        while a != b:
            if a:
                a = a.next
            else:                          # a = a.next if a else headB 
                a = headB                   
                                           
            if b:
                b = b.next
            else:                          # b = b.next if b else headA
                b = headA

        return a 
```

```python
class Solution(object):
    def BruteForce3_getIntersectionNode(self, headA, headB):
        stackA, stackB = [], []
        
        while headA:
            stackA.append(headA)
            headA = headA.next
        
        while headB:
            stackB.append(headB)
            headB = headB.next
        
        intersection = None
        while stackA and stackB and stackA[-1] == stackB[-1]:
            intersection = stackA.pop()
            stackB.pop()
        
        return intersection 
```

```python
class Solution(object):
    def findKthNode(self,head,k):
        temp=head
        while temp and k>1:
            temp=temp.next
            k=k-1
        return temp

    def reverseList(self, head):
        temp_pointer = head
        prev = None
        new_head=None

        while temp_pointer:
            new_head=temp_pointer.next
            temp_pointer.next=prev
            prev=temp_pointer
            temp_pointer=new_head
        return prev
    
    def reverseKGroup(self, head, k):
        if not head or k == 1:
            return head
        temp_pointer=head
        prev_node=None
        
        while temp_pointer:
            kth_node=self.findKthNode(temp_pointer,k)
            if kth_node==None:
                if prev_node:
                    prev_node.next=temp_pointer
                break
            next_node=kth_node.next
            kth_node.next=None
            self.reverseList(temp_pointer)
            if temp_pointer==head:
                head=kth_node
            else:
                prev_node.next=kth_node
            prev_node=temp_pointer
            temp_pointer=next_node
                
        return head

        
```

```python
class Solution(object):
    def lastElement(self, head):
        temp = head
        while temp.next:
            prev = temp 
            temp = temp.next
        return prev, temp  

    def rotateRight(self, head, k):
        if not head or not head.next or k == 0:
            return head
        
        length = 1
        temp = head
        while temp.next:
            temp = temp.next
            length += 1

        k = k % length
        if k == 0:
            return head  

        new_head = head
        for _ in range(k):
            prev, last_node = self.lastElement(new_head) 
            prev.next = None  # Break the link to last node
            last_node.next = new_head  # Attach last node to head
            new_head = last_node  # Update new head

        return new_head
```

```python
class Solution(object):
    def rotateRight(self,head, k):
        if not head or not head.next or k == 0:
            return head

        # Step 1: Find the length of the list
        length = 1
        tail = head
        while tail.next:
            tail = tail.next
            length += 1

        # Step 2: Optimize k
        k = k % length
        if k == 0:
            return head  # No rotation needed

        # Step 3: Find the new tail (length - k - 1)
        temp = head
        for _ in range(length - k - 1):
            temp = temp.next

        # Step 4: Rearrange pointers
        new_head = temp.next
        temp.next = None
        tail.next = head  # Connect old tail to old head

        return new_head
```

```python
class Node:
    def __init__(self, val=0, next=None, random=None):
        self.val = val
        self.next = next
        self.random = random

class Solution:
    def insertCopyInBetween(self, head):
        temp = head
        while temp:
            nextElement = temp.next
            copy = Node(temp.val)  # Create copy node
            temp.next = copy
            copy.next = nextElement
            temp = nextElement  # Move to next original node

    def connectRandomPointers(self, head):
        temp = head
        while temp:
            copyNode = temp.next
            if temp.random:
                copyNode.random = temp.random.next
            else:
                copyNode.random = None
            temp = temp.next.next  # Move to next original node

    def getDeepCopyList(self, head):
        temp = head
        dummyNode = Node(-1)
        res = dummyNode

        while temp:
            res.next = temp.next  # Extract the copied node
            res = res.next
            temp.next = temp.next.next if temp.next else None  # Restore original list
            temp = temp.next  # Move to next original node
        return dummyNode.next

    def copyRandomList(self,head):
        if not head:
            return None
        self.insertCopyInBetween(head)
        self.connectRandomPointers(head)
        return self.getDeepCopyList(head)


        
```
